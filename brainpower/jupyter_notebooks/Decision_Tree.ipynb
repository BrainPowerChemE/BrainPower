```python
import pandas as pd
import numpy as np
import sklearn.preprocessing
import sklearn.tree
```

```python
dev = pd.read_csv("dev.csv")
```

```python
test = pd.read_csv("test.csv")
```

```python
val = pd.read_csv("val.csv")
```

```python
train = pd.read_csv("train.csv")
```

```python
train.head()
```

```python
train_cols = train.loc[:, train.columns != 'group']
train_cols.head()
```

```python
val_cols = val.loc[:, val.columns != 'group']
val_cols.head()
```

```python
train_X = train_cols.values
val_X = val_cols.values

train_y_regr = train['group'].values.reshape(-1,1)
val_y_regr = val['group'].values.reshape(-1,1)
```

```python
scaler = sklearn.preprocessing.StandardScaler()
train_X = scaler.fit_transform(train_X)
val_X = scaler.transform(val_X)
```

```python
clf = sklearn.tree.DecisionTreeClassifier()

clf = clf.fit(train_X, train_y_regr)

y_pred = clf.predict(val_X)
```

```python
print("Accuracy:", sklearn.metrics.accuracy_score(val_y_regr, y_pred))
```
